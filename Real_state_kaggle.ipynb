# Handling Missing data

Removing the column having high amount of missing value and rows having lowest amount of missing term

```python
li_col=[]
li_row=[]
for k in missing[missing['NaN_percentage_train']>0.45].index.values:
    li_col.append(k)
    train.drop(columns=k,axis=1,inplace=True)# eliminating columns
for k in missing[(missing['NaN_percentage_train']<0.03)&(missing['NaN_percentage_train']!=0)].index.values:
    li_row.append(k)
    train.dropna(subset=[k],inplace=True) #eliminating rows
```

# Final Model and Making Prediction

Using ensemble Voting Technique

```python
#lnr_rgr=LinearRegression(fit_intercept = True, normalize= False, positive= False)

lsso_rgr= Lasso(alpha= 0.1,max_iter= 5000,positive= False,selection= 'random',warm_start= True)

rdg_rgr= Ridge(alpha= 0.9,max_iter= 1000, normalize= True,random_state= 42)

#knn_rgr= KNeighborsRegressor(algorithm= 'ball_tree',leaf_size= 20,n_neighbors= 5,p= 1,weights= 'distance')

dcs_tree= DecisionTreeRegressor(criterion='mse',max_depth= 95,max_features= 'auto',max_leaf_nodes= None,min_impurity_decrease= 0.0,min_impurity_split= 0.5,min_samples_leaf= 3,min_samples_split= 4,min_weight_fraction_leaf= 0.0,random_state= None,splitter='random')

elstc_rgr= ElasticNet(alpha= 0.1,l1_ratio= 0.9,max_iter= 5000,normalize=False,random_state= None,selection= 'random',tol= 0.001,warm_start= True)
```

```python
estimators = [('dcs_tree',dcs_tree),('elstc_rgr',elstc_rgr),('lsso_rgr',lsso_rgr)]
```

```python
from sklearn.ensemble import VotingRegressor
vr = VotingRegressor(estimators=estimators,n_jobs=-1) 
score = cross_val_score(vr,x_train,y_train,scoring='r2',cv=10)
print(np.round(np.mean(score),2))
```
